```python
#membaca sumber data dari GD menggunakan pandas
import pandas as pd

#perintah (pd) memabaca data csv yang tersimpan di GD folder dataset 
#kemudian menyimpannya dalam fariable misal : df untuk memudahkan menyaring data
df = pd.read_csv("/content/drive/My Drive/dataset/heart.csv")

#contoh melihat 5 data teratas dari dataset heart tersebut
df.head(5)
```

```python
#menampilkan data dari df untuk usia diatas 35 tahun dan dibawah 41 tahun dan disimpan ke fariable df2
df2 = df[(df['age']>35)&(df['age']<41)]
df2
```

```python
x = df.iloc[:, 0:13]
y = df.iloc[:, 13]
x
```

```python
#melakukan pengacakan data menggunakan sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1)

y_train
```

```python
#melakukan klasifikasi data
from sklearn.svm import SVC

clf = SVC()
clf.fit(x_train, y_train)
y_predict = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_predict, y_test))
```
