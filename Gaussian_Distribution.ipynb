```python
#distribusi Normal (Gaussian Distribution)
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt, seaborn as sns
```

```python
xa = stats.norm(4, 2).rvs(10000)
xb = stats.norm(7, 2). rvs(10000)
```

```python
sns.kdeplot(xa, color='red', shade=True)
sns.kdeplot(xb, color='green', shade=True)

#proses standarisasi deviasi 1 dan rata 0 (proses standarisasi)
x_standard = (xa - xa.mean()) / xa.std()
sns.kdeplot(x_standard, color='yellow', shade=True)
```

```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt
t = np.linspace(6, 10, 500)
w = chirp(t, f0=4, f1=2, t1=5, method='linear')
plt.plot(t, w)
plt.title("Linear Chirp")
plt.xlabel('time in sec)')
plt.show()
```
